Following are the types of functional argument
1. Required argument - We need to pass all required arguments while function calling and defining the function.
2. Keyword argument - We can change print sequence while calling the function.
3. Default argument - If we initialize parameter while defining function its not required/Mandatory to give it while calling function.

# **Generator and Iterator in Python :**

**Generators in Python :**

Building an iterator in Python requires a significant amount of effort. We must create a class containing __iter__() and __next__() methods, keep track of internal states and raise StopIteration when no values are returned. This is both long and contradictory. In such cases, the generator comes to the rescue. 

Python has a generator that allows you to create your iterator function. A generator is somewhat of a function that returns an iterator object with a succession of values rather than a single item. A yield statement, rather than a return statement, is used in a generator function. 

The difference is that, although a return statement terminates a function completely, a yield statement pauses the function while storing all of its states and then continues from there on subsequent calls.

```python
# Program to print the Power of two up to the given number
def PowerTwoGen( max=0 ):
   n = 1
   while n < max:
       yield 2 ** n
       n += 1

a = PowerTwoGen(6)

# Printing the values stored in a
for i in a:
   print(i)
```

**Iterators in Python :**

In Python, an iterator is an object used to iterate over iterable objects such as lists, tuples, dictionaries, and sets. An object is called iterable if we can get an iterator from it or loop over it. 

Let us understand this with one example:

Suppose you have a list of even numbers:

Numbers = [2, 6, 8, 10, 12]

You can efficiently process the traversing using a For or While loop to get each element one by one.

```python
# Program to print the list using Iterator protocols
X = [25, 78, 'Coding', 'is', '<3']
# Get an iterator using iter() 
a = iter(X)

# Printing the a iterator
print(a)

# next() for fetching the 1st element in the list that is 25
print(next(a))

# Fetch the 2nd element in the list that is 78
print(next(a))

# Fetching the consecutive elements
print(next(a))
print(next(a))
print(next(a))
```

# **List Comprehension :**

It is very concise way to create a new list by performing an operation on each item in the existing list.

                            Syntax : [Expression for Element in Iterable if condition ] 
                            Expression - Output expression
                            Iterable  - 

Example : [ X*X for X in range(1,6) ]
                        

```python
# range in list comprehension
[ X*X for X in range(1,6) ]
```

```python
# String in list comprehension
temp="Hello World!"
[ '$'+X for X in temp ]
```

```python
# List in list comprehension
[ a for a in [1,2,"abhi",4.5,[5]] ]
```

```python
# Nested if in list comprehension
[x for x in range(1,100) if x%2==0 if x%5==0]
```

```python
# if else in list comprehension
["Even" if x%2==0 else "Odd" for x in [1,3,4,9,5,6,11,13,16]]
```

```python
# if with and in list comprehension
temp=[1,2,3,4,5,6,7,8]
[x for x in temp if x%2==0 and x%3==0]
```

```python
temp=[1,2,3,4,5,6,7,8]
[a for a in temp if a%2==0 & a%3==0]
```

```python
# if with or in list comprehension
temp=[1,2,3,4,5,6,7,8]
[x for x in temp if x%2==0 or x%3==0]
```

# **Exception Handling in Python :**

Exception occures at runtime
Errors occures at Compile time

To handle the exception we need some methodology called "Exception Handling" by using this we are handling the errors in modern way and not allowing the coplete application to crash/close unexpectedly.

syntax:
    
          try:
              #programe code
              #code block
          except:
              #exception handle
          else:
              #execution
          finally:
              #always execute
       


```python
try:
  a=int(input("Enter 1st Number : "))
  b=int(input("Enter 2nd Number : "))
  f=a/b
except Exception as e:
  print("Please Enter Non-Zero Value !")
  print("Error : ",e)

else:
  print("Output : ",a/b)
finally:
  print("Finally Block Always Executes !")
```

```python
lst=[1,2,3,4,5,6]
print(lst[0])
try:
  print(lst[7])
except Exception as e:
  print("Error : ",e)
```

```python
try:
  a=int(input("Enter 1st Number : "))
  b=int(input("Enter 2nd Number : "))
  f=a/b
except ZeroDivisionError as s:
  print(s)
except Exception as e:
  print("Please Enter Non-Zero Value !")
  print("Error : ",e)

else:
  print("Output : ",a/b)
finally:
  print("Finally Block Always Executes !")
```

#**Unsupervised Learning :** ( Trained on Unlabeled Clustered Data )       
The algorithm scans through data sets looking for any meaningful connection. The data that algorithm train on as well as the predictions or recommendations they output are predetermined.

[K-Means Clustering :](https://static.javatpoint.com/tutorial/machine-learning/images/k-means-clustering-algorithm-in-machine-learning.png)

It is type of Unsupervised learnig which is used when you have unlabled data ( data without defining categeory or group ) the goal of this algorithm is to find groups in data with the number of groups represented by variable K.


**Types:**
1. Hierarchcal
2. Partationing

[Euclidean Distance :](https://editor.analyticsvidhya.com/uploads/43191elbow_img%20(1).png)

As discussed above, the Euclidean distance formula helps to find the distance of a line segment. Let us assume two points, such as (x1, y1) and (x2, y2) in the two-dimensional coordinate plane.

Thus, the Euclidean distance formula is given by:


                                d =√[(x2 – x1)2 + (y2 – y1)2]


Where,

“d” is the Euclidean distance

(x1, y1) is the coordinate of the first point

(x2, y2) is the coordinate of the second point.


[Elbow Method :](https://editor.analyticsvidhya.com/uploads/43191elbow_img%20(1).png)

The elbow method runs k-means clustering on the dataset for a range of values for k (say from 1-10) and then for each value of k computes an average score for all clusters. By default, the distortion score is computed, the sum of square distances from each point to its assigned center.

# **Decision Tree :**

**What is Entropy ?**

It is an Information Query Metric that measures the impurity or uncertainity in group of observations. It determines how a decision tree choose to split the data.

    E= −∑ i=1 to N pi*log2*pi

**What is Gini impurity ?**

**What is information  gain ?**

we can define information gain as measure of how much information a feature provides about the class. Information gain helps to determine the order of attribute in the nodes of decision tree.

The node having high information  gain that sequence is used for splitting. 
      
    [Entropy*n(pos)/total]+[Entropy*n(neg)/total]

if Entropy is low information  gain is high

if Entropy is high information  gain is low

last node in tree is called as leaf node.

In-between nodes are called as sub node or sub tree.
